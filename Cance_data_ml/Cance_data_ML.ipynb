```python
x = df.drop(columns='diagnosis')
y = df['diagnosis']
```

```python
num_of_cols = 4
num_of_rows = math.ceil(len(x.columns)/num_of_cols)
x.plot(kind='box', subplots=True, layout=(num_of_rows,num_of_cols), figsize=(20,4*num_of_rows))
plt.show()
```

```python
Z = (x - x.mean()) / x.std()
threshold = 3
outlier_mask = (Z.abs() > 3).any(axis=1)
```

```python
x_clean = x[~outlier_mask]
y_clean = y[~outlier_mask]
```

```python
num_of_cols = 4
num_of_rows = math.ceil(len(x.columns)/num_of_cols)
x_clean.plot(kind='box', subplots=True, layout=(num_of_rows,num_of_cols), figsize=(20,4*num_of_rows))
plt.show()
```

```python
#IQR Method for removing outliers:
Q1 = x_clean.quantile(0.25)
Q3 = x_clean.quantile(0.75)

IQR = Q3 - Q1

lower= Q1 - 1.5 * IQR
upper= Q3 + 1.5 * IQR

X = x_clean.clip(lower=lower,upper=upper,axis=1)
```

```python
num_of_cols = 4
num_of_rows = math.ceil(len(x.columns)/num_of_cols)
X.plot(kind='box', subplots=True, layout=(num_of_rows,num_of_cols), figsize=(20,4*num_of_rows))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y_clean,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
k = int(np.sqrt(len(x_train)))
k
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=7, weights='distance')
knc.fit(x_train,y_train)
```

```python
y_pred = knc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
acc_score = accuracy_score(y_test,y_pred)
class_report = classification_report(y_test, y_pred)
confu_matrix = confusion_matrix(y_test,y_pred)
print(f'Accuracy_score:{acc_score}, Classification_report:{class_report},Confusion_matrix:{confu_matrix}')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=confu_matrix, display_labels=[0,1])
disp.plot(cmap='Blues')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score,f1_score

precision = precision_score(y_test, y_pred, pos_label='B')
recall = recall_score(y_test, y_pred, pos_label = 'B')
f1 = f1_score(y_test, y_pred,pos_label='B')
print(f"Precision score: {precision*100:.2f}%")
print(f"Recall score: {recall*100:.2f}%")
print(f"F1 score: {f1*100:.2f}%")
```

```python
precision = precision_score(y_test, y_pred, pos_label='M')
recall = recall_score(y_test, y_pred, pos_label = 'M')
f1 = f1_score(y_test, y_pred,pos_label='M')
print(f"Precision score: {precision*100:.2f}%")
print(f"Recall score: {recall*100:.2f}%")
print(f"F1 score: {f1*100:.2f}%")
```
